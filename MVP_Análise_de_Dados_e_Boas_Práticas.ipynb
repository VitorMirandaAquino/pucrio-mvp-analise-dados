# Descrição do Problema

O conjunto de dados **HR Analytics** é um conjunto de dados multivariado que contém informações sobre funcionários de uma empresa, incluindo características demográficas, profissionais e comportamentais. O objetivo principal é analisar e prever a rotatividade de funcionários (attrition) com base em diversas variáveis como idade, salário, tempo de empresa, satisfação no trabalho, distância de casa, entre outras. O problema visa identificar fatores que contribuem para o desligamento de funcionários.

## Hipóteses do Problema

As hipóteses em relação ao **comportamento dos funcionários**:
- O nível de satisfação geral está baixo o que pode levar ao desligamento
- Os funcionários estão precisando viajar muito
- Os funcionários estão precisando fazer muitas horas extras
- Os funcionários estão recebendo poucos benefícios
- Os funcionários não estão conseguindo equilibrar a vida com o trabalho
- Os funcionários estão levando muito anos até serem promovidos
- Os funcionários estão levando muito tempo até chegar ao trabalho

As hipóteses que levam ao **desligamento da empresa**:
- Os funcionários mais novos estão se desligando mais da empresa por um comportamento da nova geração
- Pessoas que precisam de um deslocamento maior para ir trabalhar, também são mais propensas a deixar empresa
- Pessoas com salário menor são propensas a desligar da empresa
- Pessoas com mais tempo de empresa são menos propensas a deixar empresa
- Pessoas que realizam horas extras estão mais propensas a desligar da empresa

## Tipo de Problema

Este é um problema de **classificação supervisionada**. Dado um conjunto de características, o objetivo é prever se o funcionário irá deixar a empresa.

## Seleção de Dados

O dataset de Recursos humanos foi obtido no kaggle e contém diversas informações sobre funcionários de algumas empresas e se eles deixaram a empresa ou não.
```python
def analise_univariada_categorica(df: pd.DataFrame, coluna: str):
    """
    Função para analisar a distribuição de uma variável categórica
    Args:
        df: dataframe com os dados
        coluna: coluna a ser analisada
    """
    print(f"Análise univariada da coluna {coluna}")
    print(f"Número de valores únicos: {df[coluna].nunique()}")
    
    # Calcula as contagens e ordena da menor para a maior
    contagens = df[coluna].value_counts(normalize=True).sort_values()

    # Calcula o percentual de cada valor
    percentuais = (contagens * 100).round(2).reset_index()
    percentuais.columns = [coluna, "Percentual"]
    display(percentuais)
    
    # Cria um gráfico de barras para visualizar a distribuição dos valores
    plt.figure(figsize=(20, 6))
    sns.countplot(x=coluna, data=df, order=contagens.index, color="blue")
    plt.title(f"Distribuição da coluna - {coluna}", fontsize=14, fontweight="bold", color='blue')
    plt.xlabel(coluna, fontsize=11, color='black')
    plt.ylabel("Frequência", fontsize=11, color='black')
    sns.despine()
    plt.tight_layout()
    plt.show()
    
    
```

```python
def analise_univariada_numerica(df: pd.DataFrame, coluna: str):
    """
    Função para analisar a distribuição de uma variável numérica
    Args:
        df: dataframe com os dados
        coluna: coluna a ser analisada
    """
    print(f"Análise univariada da coluna {coluna}")
    print(f"Número de valores únicos: {df[coluna].nunique()}")
    
    # Estatísticas descritivas básicas
    print(f"\nEstatísticas descritivas:")
    print(f"Média: {df[coluna].mean():.2f}")
    print(f"Mediana: {df[coluna].median():.2f}")
    print(f"Desvio padrão: {df[coluna].std():.2f}")
    print(f"Mínimo: {df[coluna].min():.2f}")
    print(f"Máximo: {df[coluna].max():.2f}")
    
    # Tabela de percentis de 10 em 10
    print(f"\nTabela de percentis (de 10 em 10):")
    percentis = [10, 20, 30, 40, 50, 60, 70, 80, 90]
    valores_percentis = df[coluna].quantile([p/100 for p in percentis])
    
    tabela_percentis = pd.DataFrame({
        'Percentil': [f'{p}%' for p in percentis],
        'Valor': valores_percentis.values.round(2)
    })
    display(tabela_percentis)
    
    # Criar figura com histograma e boxplot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 12), sharex=True)
    
    # Histograma
    ax1.hist(df[coluna], bins=30, alpha=0.7, color='blue', edgecolor='black')
    ax1.axvline(df[coluna].mean(), color='red', linestyle='--', linewidth=2, label=f'Média: {df[coluna].mean():.2f}')
    ax1.axvline(df[coluna].median(), color='green', linestyle='--', linewidth=2, label=f'Mediana: {df[coluna].median():.2f}')
    ax1.set_ylabel('Frequência', fontsize=11, color='black')
    ax1.legend()
    
    # Boxplot
    ax2.boxplot(df[coluna], vert=False, patch_artist=True, 
                boxprops=dict(facecolor='blue', alpha=0.7),
                medianprops=dict(color='red', linewidth=2),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'),
                flierprops=dict(marker='o', markerfacecolor='red', markersize=4))
    ax2.set_xlabel(coluna, fontsize=11, color='black')
    ax2.set_yticks([])
    
    # Adicionar estatísticas no boxplot
    stats_text = f'Q1: {df[coluna].quantile(0.25):.2f}\nQ3: {df[coluna].quantile(0.75):.2f}\nIQR: {df[coluna].quantile(0.75) - df[coluna].quantile(0.25):.2f}'
    ax2.text(0.95, 0.98, stats_text, transform=ax2.transAxes, fontsize=10,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    fig.suptitle(f'Análise univariada da coluna {coluna}', fontsize=14, fontweight='bold', color='blue')
    plt.tight_layout()
    plt.show()
    
    # Detectar outliers usando IQR
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    print(f"\nAnálise de Outliers (método IQR):")
    print(f"Limite inferior: {limite_inferior:.2f}")
    print(f"Limite superior: {limite_superior:.2f}")
    print(f"Número de outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}% dos dados)")
```

```python
def cramers_v(confusion_matrix):
    """
    Calcula o Cramér's V dado uma matriz de contingência.
    
    Args:
        confusion_matrix (np.array): Matriz de contingência

    Returns:
        float: Valor de Cramér's V
    """
    chi2 = chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

def analise_associacao(df: pd.DataFrame, lista_colunas: list):
    """
    Função para análise de associação entre variáveis categóricas

    Args:
        df (pd.DataFrame): DataFrame com os dados
        lista_colunas (list): Lista com os nomes das variáveis
    """
    matriz = pd.DataFrame(index=lista_colunas, columns=lista_colunas, dtype=float)

    for col1 in lista_colunas:
        for col2 in lista_colunas:
            if col1 == col2:
                matriz.loc[col1, col2] = 1.0
            else:
                tabela = pd.crosstab(df[col1], df[col2])
                valor_cramer = cramers_v(tabela.values)
                matriz.loc[col1, col2] = valor_cramer

    # Criar máscara para ocultar a parte superior da matriz
    mask = np.triu(np.ones_like(matriz, dtype=bool))

    # Visualização da matriz
    plt.figure(figsize=(10, 8))
    sns.heatmap(matriz.astype(float), mask=mask, annot=True, fmt=".2f", cmap="YlGnBu", vmin=0, vmax=1)
    plt.title("Matriz de Associação entre Variáveis Categóricas (Cramér's V)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
def analise_correlacao(df: pd.DataFrame, lista_colunas: list, metodo: str = "spearman"):
    """
    Função para análise de correlação de variaveis numéricas

    Args:
        df (pd.DataFrame): DataFrame com os dados
        lista_colunas (list): Lista com nomes das colunas numéricas
        metodo (str): Tipo de correlação ('pearson', 'spearman', 'kendall')
    """
    # Calcula a matriz de correlação
    corr_matrix = df[lista_colunas].corr(method=metodo)

    # Máscara para exibir apenas a parte inferior
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Visualização
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
                cmap="coolwarm", vmin=-1, vmax=1, square=True,
                cbar_kws={"shrink": 0.8})
    
    plt.title(f'Matriz de Correlação ({metodo})')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
def analise_bivariada_categoricas(df: pd.DataFrame, variavel_comparativa: str, variavel_alvo: str):
    """ 
    Função para análise de duas variáveis categóricas
    Faz um gráfico de barras empilhadas (100%) mostrando a proporção das categorias da variável alvo
    para cada categoria da variável comparativa.

    Args:
        df (pd.DataFrame): DataFrame com os dados
        variavel_comparativa (str): Nome da variável categórica comparativa
        variavel_alvo (str): Nome da variável alvo 
    """
    # Converte em string caso a variável comparativa seja numérica
    if df[variavel_comparativa].dtype == 'int64' or df[variavel_comparativa].dtype == 'float64':
        df[variavel_comparativa] = df[variavel_comparativa].astype(str)
        
    # Calcular proporções para obter os valores percentuais
    prop = pd.crosstab(df[variavel_comparativa], df[variavel_alvo], normalize='index') * 100
    
    # Criar figura
    plt.figure(figsize=(12, 8))
    
    # Gráfico de barras empilhadas usando seaborn
    ax = sns.histplot(data=df, x=variavel_comparativa, hue=variavel_alvo, 
                     multiple="fill", shrink=.8, palette=['red', 'blue'])
    
    # Configurações do gráfico
    plt.ylabel('Percentual')
    plt.xlabel(variavel_comparativa)
    plt.title(f'Distribuição percentual de "{variavel_alvo}" por "{variavel_comparativa}"')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 1)
    
    # Converter eixo y para percentual
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    
    # Obter as categorias na ordem que aparecem no gráfico
    categorias_grafico = [tick.get_text() for tick in ax.get_xticklabels()]
    
    # Adicionar valores percentuais dentro das barras
    for i, categoria in enumerate(categorias_grafico):
        # Obter as proporções para esta categoria
        valores = prop.loc[categoria]
        
        # Posição inicial para o texto
        acumulado = 0
        
        # Ordenar os valores na mesma ordem que aparecem no gráfico
        # O seaborn geralmente coloca a primeira categoria do hue no fundo
        categorias_alvo = sorted(valores.index)
        
        for j, valor_cat in enumerate(categorias_alvo):
            percentual = valores[valor_cat]
            
            if percentual > 5:  # Só mostra se for maior que 5% para não sobrepor
                # Posição do texto (meio da seção da barra)
                pos_texto = acumulado + (percentual / 200)  # dividir por 200 porque ylim é 0-1
                
                # Cor do texto (branco para melhor contraste)
                plt.text(i, pos_texto, f'{percentual:.1f}%', 
                        ha='center', va='center', fontweight='bold', 
                        color='white', fontsize=10)
            
            acumulado += percentual / 100  # dividir por 100 porque ylim é 0-1
    
    plt.tight_layout()
    plt.show()
    
    # Exibir também a tabela de percentuais
    print(f"\nTabela de percentuais (% por linha):")
    display(prop.round(2))
```

### Correlação

| Valor de r (em módulo) | Intensidade da correlação     |
|------------------------|--------------------------------|
| 0,00 < 0,20            | Muito fraco / quase sem relação |
| 0,20 < 0,40            | Fraco                          |
| 0,40 < 0,60            | Moderado                       |
| 0,60 < 0,80            | Forte                          |
| 0,80 < 1,00            | Muito forte                    |

Fonte: https://www.statstutor.ac.uk/resources/uploaded/pearsons.pdf

```python
# Criar uma cópia do dataframe para análise de correlação
colunas_correlacao = [
    'Attrition', 'Age', 'DistanceFromHome', 'EmployeeNumber', 'MonthlyIncome', 'NumCompaniesWorked',
    'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany',
    'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager'
]
df_para_correlacao = df[colunas_correlacao].copy()
df_para_correlacao['Attrition'] = df_para_correlacao['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
analise_correlacao(
    df_para_correlacao,
    colunas_correlacao,
    metodo = 'pearson'
)
```

**Correlações relevantes**
- As variáveis de ano em determinada situação (Empresa, Mesmo gerente, última promoção e no mesmo cargo) estão com correlações fortes entre si e não devemos seguir com todas para um possível modelo
- A variável de total de anos trabalhando também tem um correlação moderado para forte com as variáveis de ano em determinada situação
- A variável de renda mensal tem uma correlação moderada até forte com variáveis relacionadas a maior tempo de trabalho ou de vida

**Correlações com a variável alvo**
